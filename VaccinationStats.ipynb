# **Question 4:** Implement code to fill the missing data.

````python
# We import our data set into a pandas DataFrame to easily work with.
data_set = pd.read_csv("https://www.piworks.net/Upload/Document/Original/country_vaccination_stats.csv")
data_set
````

````python
countries = data_set.country.unique()#We are extracting the countries.
for i in range(0,60):#Iterating through.
  country_based = data_set.loc[data_set['country'] == countries[i]]#Getting that country related datas.
  val = country_based['daily_vaccinations'].min()#Finding the minimum value of that relevant country.
  if np.isnan(val):#If that country has no valid vaccination number then we should insert 0.
    val = 0
  index = country_based.iloc[0].name#Row number of the first row of that country.
  data_set.iloc[index,2] = val#Insert operation.
````

# **Question 5:** Implement code to list the top-3 countries with highest median daily vaccination numbers by considering missing values imputed version of dataset

````python
#We are creating a dictionary to hold country-median pairs.
medians = {}
````

````python
for i in range(0,60):#Iterating trough countries.
  country_based = data_set.loc[data_set['country'] == countries[i]]#Getting that country related datas.
  medians[countries[i]] = country_based.median(numeric_only=True).iloc[0]#We are adding country-median pairs into our dictionary.
````

````python
#We are sorting our dictionary descending by our values then listing the top 3 countries.
sorted_medians = sorted(medians.items(), key=lambda x:x[1], reverse=True)
sorted_medians[0],sorted_medians[1],sorted_medians[2]
````
